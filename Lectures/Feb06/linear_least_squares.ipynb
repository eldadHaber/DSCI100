```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# data
x = np.array([1,   2, 3, 4.0])
y = np.array([1.0, 4, 9, 16])

# setup the matrix
A = np.vstack((x, np.ones(len(x)))).T
b = y

print(A)
print(b)
```

```python
# Solve for the coefficients
output = np.linalg.lstsq(A, b, rcond=None)

theta = output[0]
r     = output[1]

print('theta = ', theta)

print('r = ', r)

residuals = b - A@theta

print('residuals = ', residuals)
```

```python
bpred = A@theta

# Visualize
plt.plot(x, y, 'o')
plt.plot(x, bpred)
plt.show()
```

```python
x = np.array([-16, -7, -2, 1, 2, 3, 4, 5.0])
y = np.array([22, 40, 6, 1, 4.7, 9.7, 15, 30])

A = np.vstack((x**2, x, np.ones(len(x)))).T
b = y

output = np.linalg.lstsq(A, b, rcond=None)
theta = output[0]
r     = output[1]

print('theta = ', theta)
print('r = ', r)

bpred = A@theta

# Visualize
plt.plot(x, y, 'o')
plt.plot(x, bpred)
plt.show()
```
